6\. Use broadcasting to create a grid of distances

Route 66 crosses the following cities in the US: Chicago, Springfield, Saint-Louis, Tulsa, Oklahoma City, Amarillo, Santa Fe, Albuquerque, Flagstaff, Los Angeles.
The corresponding positions in miles are: 0, 198, 303, 736, 871, 1175, 1475, 1544, 1913, 2448.

  * Construct a 2D grid of distances among each city along Route 66
  * Convert that in km (those savages...)

````python
cities = np.array(["Chicago", "Springfield", "Saint-Louis", "Tulsa", "Oklahoma City", "Amarillo", "Santa Fe", "Albuquerque", "Flagstaff", "Los Angeles"])

miles = np.array([0, 198, 303, 736, 871, 1175, 1475, 1544, 1913, 2448])

kms = miles*1.6

# np.tile(miles,(miles.shape[0],1)) repeats miles vertically (each row is a miles array)
# np.tile(miles,(miles.shape[0],1)).T returns miles horizontally repeated (each columns is miles array)

miles_grid = np.abs(np.tile(miles,(miles.shape[0],1)).T - np.tile(miles,(miles.shape[0],1)))

km_grid = miles_grid*1.6

print('Distances grid (miles): \n', miles_grid, '\n')
print('Distances grid (km):\n', km_grid)
````

````python
# fancy visualization

fig, ax = plt.subplots(figsize=(8,8))
im = ax.imshow(miles_grid)

ax.set_xticks(np.arange(len(cities)))
ax.set_yticks(np.arange(len(cities)))
ax.set_xticklabels(cities)
ax.set_yticklabels(cities)

plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

threshold = im.norm(miles_grid.max())/2.

textcolors = ["white", "black"]
for i in range(len(cities)):
    for j in range(len(cities)):
        text = ax.text(j, i, miles_grid[i, j], ha="center", va="center", color=textcolors[int(im.norm(miles_grid[i, j]) > threshold)])
        
ax.set_title("Distance between Route66 cities (in miles)", fontsize=20)

fig.tight_layout()

plt.show()
````

7\. Prime numbers sieve: compute the prime numbers in the 0-N (N=99 to start with) range with a sieve (mask).
  * Construct a shape (100,) boolean array, the mask
  * Identify the multiples of each number starting from 2 and set accordingly the corresponding mask element
  * Apply the mask to obtain an array of ordered prime numbers
  * Check the performances (timeit); how does it scale with N?
  * Implement the optimization suggested in the [sieve of Eratosthenes](https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes)

````python
def calculate_primes(N):
    numbers = np.arange(2, N, 1)
    mask = np.zeros_like(numbers, dtype=bool)

    start_time = time()

    for i, num in enumerate(numbers):
        primes = (mask == 1)

        # case num = 2
        if np.all(primes == 0): 
            mask[i] = 1
            continue # restarting loop

        # case num not a prime number
        if np.any((num % numbers[primes]) == 0): 
            continue
        
        # case num = prime number
        mask[i] = 1

    computation_time = time() - start_time

    return numbers[mask], computation_time
````

````python
primes, t = calculate_primes(100)
print('Prime numbers up to 100: \n', primes)
print(f'Corresponding computing time with standard algorithm: {t:.1e} s')
````

````python
# Erastothenes (O(nloglogn))

def Erastothenes(n):
    numbers = np.arange(0,n,1)
    sieve = np.ones_like(numbers, dtype=bool)

    start_time = time()

    for i in range(2, int(np.sqrt(n))):
        if sieve[i]:
            j_array = np.array([(i**2 + i*m) for m in range(0, n)])
            for j in j_array[j_array<n]:
                sieve[j] = 0
    
    computation_time = time() - start_time

    return numbers[sieve], computation_time
````

````python
primes_e, t_e = Erastothenes(100)
print('Prime numbers up to 100: \n', primes_e[2:])
print(f'Corresponding computing time with Erastothenes algorithm: {t_e:.1e} s')
print(f'Difference in computation time between the two algorithms: {abs(t-t_e):.1e} s')
````

````python
# how does it scale with N

N = np.arange(100, 6000, 100)

primes, comp_times = [], []

p_er, comp_times_er = [], []

for n in N:
    p,t = calculate_primes(n)
    pe, te = Erastothenes(n)
    primes.append(p)
    comp_times.append(t)
    p_er.append(pe)
    comp_times_er.append(te)

fig, ax = plt.subplots(figsize = (10,8))

ax.grid(alpha = 0.3)

ax.plot(N, comp_times, label='standard')
ax.plot(N, comp_times_er, label = 'Erastothenes ~ O(log(logN))')
ax.legend(title = 'Algorithm')
ax.set_xlabel('$N_{max}$')
ax.set_ylabel('computation time')
ax.set_title('Computation time for the two algorithms as a function of N')

plt.show()
````

8\. Diffusion using random walk

Consider a simple random walk process: at each step in time, a walker jumps right or left (+1 or -1) with equal probability. The goal is to find the typical distance from the origin of a random walker after a given amount of time. 
To do that, let's simulate many walkers and create a 2D array with each walker as a row and the actual time evolution as columns

  * Take 1000 walkers and let them walk for 200 steps
  * Use randint to create a 2D array of size walkers x steps with values -1 or 1
  * Build the actual walking distances for each walker (i.e. another 2D array "summing on each row")
  * Take the square of that 2D array (elementwise)
  * Compute the mean of the squared distances at each step (i.e. the mean along the columns)
  * Plot the average distances (sqrt(distance\*\*2)) as a function of time (step)
  
Did you get what you expected?

````python
walkers = 1000
steps = 200

# grid (shape = walkers x steps)
grid = np.random.randint(2,size=(walkers, steps))
grid = np.where(grid == 0, -1, 1)

# walking distance (distance walked at each step) --> squared
squared_dist = np.cumsum(grid, axis=1)**2

# average on walkers
av_dist = np.mean(squared_dist, axis=0)**(1/2)
````

````python
# plot average distance

fig, ax = plt.subplots(figsize=(8,6))

ax.plot(np.arange(steps), av_dist, label = 'simulation')
ax.plot(np.arange(steps), np.sqrt(np.arange(steps)), label = 'expected result')
ax.legend()
ax.set_title('Random Walk')
ax.set_xlabel('step')
ax.set_ylabel('average distance')
plt.show()
````

````python
data = np.loadtxt('populations.txt')
year, hares, lynxes, carrots = data.T
````

````python
h_mean = np.mean(hares)
h_std = np.std(hares)
l_mean = np.mean(lynxes)
l_std = np.std(lynxes)
c_mean = np.mean(carrots)
c_std = np.std(carrots)

fig,ax = plt.subplots(figsize = (10,6))
ax.grid(alpha = 0.5)

ax.plot(year, hares, lw = 3, label = "hares", color = "#ddaadd")
ax.plot(year, lynxes, lw = 3, label = "lynxes", color = "#99cccc")
ax.plot(year, carrots, lw = 3, label = "carrots", color = "#fe995c")

ax.axhline(h_mean, lw = 2, color = "#ddaadd", linestyle = 'dotted')
ax.axhline(l_mean, lw = 2, color = "#99cccc", linestyle = 'dotted')
ax.axhline(c_mean, lw = 2, color = "#fe995c", linestyle = 'dotted')

ax.fill_between(year, h_mean+h_std, h_mean-h_std, color = "#ddaadd", alpha = 0.2)
ax.fill_between(year, l_mean+l_std, l_mean-l_std, color = "#99cccc", alpha = 0.2)
ax.fill_between(year, c_mean+c_std, c_mean-c_std, color = "#fe995c", alpha = 0.2)

ax.set_xlabel("years")
ax.set_ylabel("popoulation numerosity")

ax.set_ylim(-2000,85000)
ax.set_xlim(1900,1920)
ax.set_xticks(np.arange(1900,1921,5))

ax.legend()
plt.show()
````
